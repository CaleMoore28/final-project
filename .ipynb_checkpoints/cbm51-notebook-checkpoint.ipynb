```python
#Import the data and external methods, show the first 10
import pandas as pd
import numpy as np

data = pd.read_csv("crime.tsv", index_col="_id", sep='\t')

data.head(10)
```

```python
#Drop the irrelevant columns and rows with a null value
newData = data.reset_index()
newData = newData.drop(columns=['_id', 'PK', 'CCR', 'HIERARCHY', 'INCIDENTLOCATION', 'INCIDENTZONE', 'INCIDENTTRACT', 'COUNCIL_DISTRICT', 'PUBLIC_WORKS_DIVISION', 'X', 'Y', 'OFFENSES'])
newData = data.dropna()

newData.head(10)
```

```python
#Remove any of the incidents that took place before January 1, 2013 (only looking at last 10 years)
for index, row in newData.iterrows():
    #Get the string value of the current row
    timeString = row['INCIDENTTIME']
    
    positionofSymbol = timeString.find("-")   #Finds the position of the dash and assign it to an integer value called positionOfSymbol
    timeString = timeString[:positionofSymbol]   #Reassign timeString to the value of the string before the first dash

    intVal = int(timeString)   #Turns the edited string value into a int value and assigns it to intVal
```
